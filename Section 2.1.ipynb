#### a. Write a Python script to read the sheet and perform the following analysis:
####   - Read the ‘Description’ of the companies and create a extract only the services or product offerings from each company in a different column.


```python
import re
import pandas as pd
```

```python
df = pd.read_excel('Trendlytics.xlsx', sheet_name='funding_data')
```

```python
df.duplicated()
```

```python
def extract_services(description):
    # Define keywords to look for
    keywords = [
        'retail', 'products', 'goods', 'merchandise', 'items', 'consumergoods', 'shopping',
        'apparel', 'clothing', 'fashion', 'garments', 'attire', 'wardrobe', 'fashion',
        'technology', 'software', 'hardware', 'IT', 'services', 'techsolutions', 'digitalsolutions',
        'healthcare', 'medical services', 'wellness', 'health products', 'pharmaceuticals',
        'food', 'beverages', 'restaurant', 'catering', 'culinary', 'grocery',
        'finance', 'banking', 'financial services', 'investments', 'insurance', 'wealth management',
        'education', 'learning', 'training', 'educational services', 'e-learning', 'tutoring',
        'hospitality', 'hotels', 'accommodation', 'travel services', 'tourism', 'resorts',
        'automotive', 'cars', 'vehicles', 'car services', 'auto parts', 'transportation',
        'real estate', 'property', 'housing',' real estate services', 'property management',
        'legal services', 'law firm', 'attorney', 'legal advice', 'legal solutions',
        'consulting', 'advisory services', 'management consulting', 'business consulting',
        'marketing', 'advertising', 'digital marketing', 'media services', 'branding',
        'manufacturing', 'production', 'industrial services', 'factory, manufacturing solutions',
        'environmental services', 'sustainability', 'eco-friendly solutions',' green products'
        'entertainment', 'media, digital entertainment', 'event management', 'entertainment services',
        'logistics', 'supply chain', 'distribution', 'warehousing', 'freight services',
        'construction', 'building, engineering services', 'construction materials', 'infrastructure',
        'fitness', 'sports', 'gym', 'sports equipment',' fitness services', 'athletic gear',
        'art', 'design', 'creative services', 'graphic design', 'interior design',
        'retailer', 'products', 'shopping experience', 'apparel', 'franchise',
        'information collection', 'processing service', 'delivery', 'Businessfotografie',
        'Corporate Portraits', 'application', 'solutions', 'platform', 'library', 'services',
        'systems', 'app', 'tool', 'software', 'aggregation platform', 'solutions', 'products',
        'app', 'bakery', 'breads', 'confectionary'
    ]
    
    pattern = re.compile(r'\b(?:' + '|'.join(keywords) + r')\b', re.IGNORECASE)
    
    # Find all matches in the description
    matches = pattern.findall(description)
    
    # Return the matches as a string
    return ', '.join(matches)
```

```python
df['Services/Products'] = df['Description'].apply(extract_services)
```

```python
df.to_excel('Section 2.1.xlsx', index=False)
```

####   - Determine the most funded month.

```python
df.dtypes
```

```python
df['Last Funding Date'] = pd.to_datetime(df['Last Funding Date'], format='%d-%m-%y', errors='coerce')
```

```python
df['Last Funding Month'] = df['Last Funding Date'].dt.to_period('M')
```

```python
df
```

```python
most_funded_month =df.groupby('Last Funding Month').size().reset_index(name='Funding Count')
```

```python
summary = f"""
Key Findings:
1. The most funded month is {most_funded_month}.
"""

with open('summary_report.txt', 'w') as file:
    file.write(summary)
```
