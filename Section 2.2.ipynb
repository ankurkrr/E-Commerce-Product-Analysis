```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
from selenium.common.exceptions import NoSuchElementException
```

```python
chromedriver_path = 'chromedriver\chromedriver.exe'
```

```python
options = Options()
options.add_argument('--headless')
options.add_argument('--no-sandbox')
options.add_argument('--disable-dev-shm-usage')
options.add_argument('--disable-gpu')
options.add_argument('--disable-extensions')
options.add_argument('--disable-infobars')
options.add_argument('--remote-debugging-port=9222')
service = Service(chromedriver_path)
driver = webdriver.Chrome(service=service, options=options)
```

```python
url = 'https://www.findfunding.vc'
```

```python
driver.get(url)
```

```python
time.sleep(10)
```

```python
def click_see_more(driver, wait, retries=3):
    for i in range(retries):
        try:
            see_more_button = wait.until(EC.element_to_be_clickable(
                (By.XPATH, "//button[contains(@class, 'MuiButtonBase-root') and contains(text(), 'see more')]")
            ))
            see_more_button.click()
            time.sleep(15)
            return True
        except Exception as e:
            print(f"Attempt {i+1} failed: {e}")
            time.sleep(15)
    return False
```

```python
wait = WebDriverWait(driver, 10)
```

```python
while True:
    success = click_see_more(driver, wait)
    if not success:
        print("No more 'See More' button found or an error occurred")
        break
```

```python
page_source = driver.page_source
soup = BeautifulSoup(page_source, 'html.parser')
```

```python
link_tags = soup.find_all('a', class_='MuiBox-root MuiBox-root css-1djwytu', href=True)
```

```python
link_tags
```

```python
links = []
for link in link_tags:
    links.append(link.get('href'))
```

```python
links
```

```python
name = []
contact = []
funds = []

for link in links:
    driver.get(link)
    
    time.sleep(10)

    linked_page_source = driver.page_source
    linked_soup = BeautifulSoup(linked_page_source, 'html.parser')
    investors_name = linked_soup.find_all('p',class_='sw-font-size-xs sw-text-color-000000 sw-font-family-default sw-font-weight-normal sw-text-align-left sw-letter-spacing-normal sw-padding-top-none sw-padding-bottom-5xs sw-padding-left-none sw-padding-right-none MuiBox-root css-927vqh')
    contacts = linked_soup.find_all('a',class_='email-field sw-font-size-xs sw-text-color-000000 sw-font-family-default sw-font-weight-normal sw-text-align-left sw-letter-spacing-normal sw-padding-top-none sw-padding-bottom-4xs sw-padding-left-none sw-padding-right-none')
    funding_size = linked_soup.find_all('div',class_='MuiGrid-root MuiGrid-item MuiGrid-grid-xs-12 MuiGrid-grid-sm-6 MuiGrid-grid-md-4 css-s1o41q')[6].div.div.div.div.div.div
                                        
    for n in investors_name:
        name.append(n.get_text(strip=True))
        
    for c in contacts:
        contact.append(c.get_text(strip=True))
        
    for f in funding_size:
        funds.append(f.get_text(strip=True))
```

```python
df = pd.DataFrame({'Investors Name':name,'Contacts':contact,'Funding Size':funds})
```
